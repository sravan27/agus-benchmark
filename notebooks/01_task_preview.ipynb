# AGUS Task Preview

Inspect generated AGUS tasks across families.

```python
import json
from pathlib import Path
from pprint import pprint

import pandas as pd

project_root = Path.cwd().parent
tasks_path = project_root / 'data' / 'generated' / 'agus_v1_all.json'
tasks = json.loads(tasks_path.read_text())
print(f'Loaded {len(tasks)} tasks from {tasks_path}')
```

```python
rows = []
for task in tasks:
    rows.append(
        {
            'task_id': task['task_id'],
            'family': task['family'],
            'difficulty': task['difficulty'],
            'shift_type': task['shift_type'],
            'distractor_level': task['distractor_level'],
            'num_examples': len(task['examples']),
        }
    )

summary = pd.DataFrame(rows)
summary.groupby(['family', 'difficulty']).size().reset_index(name='count')
```

```python
for family in sorted(summary['family'].unique()):
    sample = next(task for task in tasks if task['family'] == family)
    print('=' * 80)
    print(family)
    pprint(sample)
    print()
```
